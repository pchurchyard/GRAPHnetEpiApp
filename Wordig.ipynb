```python
from rpy2.robjects.packages import importr
utils = importr("utils")
import csv
import pandas as pd
import os
from os import listdir as os
from os.path import isfile, join
import glob
import warnings
import re
from collections import defaultdict
warnings.simplefilter('ignore')
```

```python
# Define path for data
PATH = "/Users/bwicht/Desktop/"

list_raw = []
list_clean = []

# Create lists of CSV Files
for filepath in glob.iglob(PATH + "DATA/Raw/*.csv"):
    list_raw.append(filepath)

for filepath in glob.iglob(PATH + "DATA/Clean/*.csv"):
    list_clean.append(filepath)
```

```python
# Store All the CSV in lists of Dataframes
master_raw = []
master_clean = []
master_full = []
country_list = []
country_list_full = []
date_list_full = []
csv_dic = {}
defaults_keys = {'date': '','csv':''}
D = dict.fromkeys(country_list, defaults_keys)




for i in range(len(list_raw)):
    country = list_raw[i][31:-19]
    country_list_full.append(country)
    date = re.search(r'[0-9]{4}_[0-9]{2}_[0-9]{2}', list_raw[i])
    date = date.group()
    date_list_full.append(date)

    if country not in country_list:
        country_list.append(country)
        key = country
        if key not in csv_dic:  
            csv_dic[country] = []
    csv_dic[country].append(date)
            
    try:
        data_file_raw = pd.read_csv(list_raw[i],sep = ";",low_memory=False,error_bad_lines=True)
        master_raw.append(data_file_raw)
    except:
        data_file_raw = pd.read_csv(list_raw[i],sep = ",",low_memory=False,error_bad_lines=True)
        master_raw.append(data_file_raw)
        
for i in range(len(list_clean)):
    try:
        data_file_clean = pd.read_csv(list_clean[i],sep = ";",low_memory=False,error_bad_lines=True)
        master_clean.append(data_file_clean)
    except:
        data_file_clean = pd.read_csv(list_clean[i],sep = ",",low_memory=False,error_bad_lines=False)
        master_clean.append(data_file_clean)
        
### Correction of miscategorized files

anomalies = [28]

for i in anomalies:
        data_file = pd.read_csv(list_raw[i],sep = ",",low_memory=False,error_bad_lines=False)
        master_raw[i] = data_file

# List containting all the pairs of document
master_full = master_raw + master_clean

D = dict.fromkeys(country_list, defaults_keys)
```
